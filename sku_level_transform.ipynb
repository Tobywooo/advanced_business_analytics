***Notebook to create SKU specific data for top 10 products***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Define the local folder filename
csv_filename = 'cltpipe_homedepot_data.csv'
# Read the CSV file from the specified path
full_path = os.path.join(r'C:\Users\tobia\OneDrive\Desktop\School\advanced_business_analytics\data', csv_filename)
df = pd.read_csv(full_path)
df.head()
```

```python
df.count()
```

```python
df_clean = df.drop(columns=['state territory code', 'sales $'])
df_clean.head(1)
```

```python
df_clean['month_year'] = df_clean['month'].str[-4:]
df_clean['month_number'] = df_clean['month'].str.extract(r'Fiscal Period (\d+)').astype(int)
df_clean = df_clean.drop(columns=['month'])
df_clean['month'] = (df_clean['month_number'].astype(str) + '-' + pd.to_datetime(df_clean['month_year']).dt.strftime('%b-%Y'))
df_clean['month'] = pd.to_datetime('1-' + df_clean['month_number'].astype(str) + '-' + df_clean['month_year'].astype(str), format='%d-%m-%Y')
df_clean = df_clean.dropna(subset=['sales units'])
df_clean.head(5)
```

```python
df_clean = df_clean.assign(sku=df_clean['sku name'].str.replace(r"[\",\-']", "", regex=True))
df_clean = df_clean.drop(columns=['sku name'])
df_clean['sku'] = df_clean['sku'].str.replace(' ', '_')
df_clean = df_clean.rename(columns={'sku': 'sku name'})
df_clean.head()
```

```python
df_clean['sub class'] = df_clean['sub class'].str.replace('-', '_')
df_clean['region'] = df_clean['region'].str.replace('-', '_')
```

```python
df_top_10_sku = df_clean.copy()
df_top_10_sku = df_top_10_sku['sku name'].value_counts().reset_index()
df_top_10_sku.columns = ['sku', 'count of sku']
df_top_10_sku.head(10)
```

```python
top_10_skus = df_top_10_sku['sku'].head(10).to_dict()
top_10_skus
```

```python
df_top_10_sku_data = df_clean[df_clean["sku name"].isin(top_10_skus.values())].copy()
df_top_10_sku_data = df_top_10_sku_data[["month", "sku name", "sku nbr", "sub class", "region", "sales units"]]
df_top_10_sku_data = df_top_10_sku_data.rename(columns={"sku name": "sku_name", "sku nbr": "sku_num", "sub class": "sub_class", "sales units": "sales_units"})
df_top_10_sku_data.head()
```

```python
df_top_10_sku_data.count() 
```

```python
df_masked = df_top_10_sku_data.copy()
df_masked = df_masked.drop(columns=['sub_class'])
df_masked['region'] = df_masked['region'].str[5:]
df_masked['region'] = df_masked['region'].str.replace(' ', '_')
df_masked.head()
```

```python
col_name = 'sales_units' 

noise = np.random.normal(loc=0, scale=df_masked[col_name].std() * 0.001, size=df_masked[col_name].shape[0])
df_masked[col_name] = df_masked[col_name] + noise
df_masked.head()
```

```python
sku_map = {
    '3/4_SCH_40_90_DEG_ELBOW_SXS_35_PK': 'PVC-ELB90-0.75-S40-SXS [35pk]',
    '3/4_SIDE_OUTLET_90_DEGREE_SOCKET': 'PVC-S-ELB90-0.75-SOC',
    '3/4_SCH_40_TEE_SXSXS_(20_PACK)': 'PVC-TEE-0.75-S40-SXSXS [20pk]',
    '3/4_PVC_COUPLING_DEEP_SOCKET_SXS': 'PVC-CPL-0.75-DS-SXS',
    '11/4X2FT_PVC_SCH_40_PIPE': 'PVC-PIPE-1.25-S40 [2FT]',
    '3/4_SCH_40_COUPLING_SXS_(25_PACK)': 'PVC-CPL-0.75-S40-SXS [25pk]',
    '1_PVC_SIDE_OUTLET90_DEG_ELBOW_SXSXS': 'PVC-S-ELB90-1.0-SXSXS',
    '3/4_PVC_RUNNING_TRAP_SPGXSPG': 'PVC-TRP-0.75-SPG-SPG',
    '1/2X2_FLOWGUARD_GOLD_CPVC': 'CPVC-GLD-0.5-PIPE [2FT]',
    '11/2X2_PVC_MALE_ADAPTER_MPTXS': 'PVC-ADP-1.5x2-MPT-SOC'
}

df_masked['sku_name_masked'] = df_masked['sku_name'].map(sku_map)

df_masked = df_masked.drop(columns=['sku_name'])

df_masked = df_masked.rename(columns={'sku_name_masked': 'sku_name'})

print(df_masked['sku_name'].head())
```

```python
df_masked.head()
```

```python
df_masked.count()
```

```python
df_masked = df_masked[['month', 'sku_name', 'sku_num', 'region', 'sales_units']]
df_masked.head()
```

```python
df_masked['sku_num'].unique()
```

```python
import random
unique_sku_array = df_masked['sku_num'].unique()
def transform_sku(val):
    multiplier = random.choice([2, 3])
    offset = random.randint(1, 20)
    sign = random.choice([-1, 1])
    return (val * multiplier) + (sign * offset)

sku_map = {sku: transform_sku(sku) for sku in unique_sku_array}
df_masked['sku_num'] = df_masked['sku_num'].map(sku_map)
```

```python
df_masked.head()
```

```python
df_timeGen_1 = df_masked.drop(columns=['region','sku_name'])
df_timeGen_1.head()
```

```python
df_timeGen_1 = df_timeGen_1.groupby(['month', 'sku_num'], as_index=False)['sales_units'].sum()
```

```python
df_timeGen_1.head()
```

```python
df_wide = df_timeGen_1.pivot(index='sku_num', columns='month', values='sales_units')
df_wide = df_wide.reset_index()
df_wide = df_wide.rename(columns={'sku_num': 'Part number'})
df_wide.columns.name = None

print(df_wide)
```

```python
df_wide.to_csv('data/first_upload_home_depot_clt_pipe.csv', index=False)
```
