```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
# Define the local folder filename
csv_filename = 'Queryclassproject_Full_2026-04-02.csv'
# Read the CSV file from the specified path
full_path = os.path.join(r'data', csv_filename)
df = pd.read_csv(full_path)
df.head()
```

```python
df['month_year'] = df['month'].str[-4:]
df['month_number'] = df['month'].str.extract(r'Fiscal Period (\d+)').astype(int)
df_clean = df.drop(columns=['month'])
df_clean['month'] = (df_clean['month_number'].astype(str) + '-' + pd.to_datetime(df_clean['month_year']).dt.strftime('%b-%Y'))
df_clean['month'] = pd.to_datetime('1-' + df_clean['month_number'].astype(str) + '-' + df_clean['month_year'].astype(str), format='%d-%m-%Y')
df_clean = df_clean.dropna(subset=['sales units'])
df_clean = df_clean.drop(columns = ['sales $','month_number','month_year',])
df_clean.head(5)
```
